## Theoretical Explanation

Matrix rank measures how many independent directions of information exist in a matrix. Rank connects linear independence, invertibility, column space, and null space. The determinant is defined only for square matrices and becomes zero when the matrix is singular.

The Frobenius norm summarizes the magnitude of all matrix entries. The trace sums diagonal entries and connects to eigenvalues later. These quantities appear in regularization, optimization, and numerical diagnostics.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import norm, matrix_rank

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 1. Frobenius Norm and Trace

```python
A = np.random.randn(4,5)
frob_direct = np.sqrt(np.sum(A**2))
frob_np = norm(A, 'fro')
trace_method = np.sqrt(np.trace(A.T @ A))
print('Frobenius direct:', frob_direct)
print('Frobenius NumPy:', frob_np)
print('Trace method:', trace_method)
```

### Output Interpretation

All three methods agree. The trace method works because the diagonal of `A.T @ A` contains column dot products with themselves.

## 2. Rank and Linear Dependence

```python
B = np.array([[1,2,3], [2,4,6], [1,0,1]])
print('Matrix B:\n', B)
print('Rank of B:', matrix_rank(B))
print('Max possible rank:', min(B.shape))
print('Determinant:', np.linalg.det(B))
```

### Output Interpretation

The matrix is rank-deficient because one row is a multiple of another. Its determinant is zero or numerically close to zero, indicating singularity.

## 3. Column Space Test by Rank Augmentation

```python
A = np.array([[1,0], [0,1], [1,1]])
v_in = np.array([[2], [3], [5]])
v_out = np.array([[2], [3], [6]])
for name, v in [('v_in', v_in), ('v_out', v_out)]:
    augmented = np.hstack([A, v])
    print(name, 'rank A =', matrix_rank(A), 'rank augmented =', matrix_rank(augmented), 'in column space?', matrix_rank(A)==matrix_rank(augmented))
```

### Output Interpretation

If augmenting the matrix with the vector does not increase rank, the vector is already in the column space. If rank increases, the vector adds a new direction.
