## Learning Objectives

This notebook covers full inverse, one-sided inverse, pseudoinverse, invertibility conditions, numerical stability, and geometric interpretation.

## Theoretical Explanation

The inverse of a matrix reverses a linear transformation and solves equations of the form `Ax=b`. A full inverse exists only for square full-rank matrices. Tall or wide matrices may have one-sided inverses when they have maximum possible rank. Every matrix has a Moore-Penrose pseudoinverse.

In applied work, explicit inversion should be used carefully because ill-conditioned matrices amplify numerical error. Solvers and decompositions are often preferred.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import norm, matrix_rank

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 1. Full Inverse

```python
A = np.array([[2,1], [1,3]])
Ainv = np.linalg.inv(A)
print('A inverse:\n', Ainv)
print('A @ Ainv:\n', np.round(A @ Ainv, 6))
```

### Output Interpretation

The product is the identity matrix, confirming that `Ainv` reverses the transformation represented by `A`.

## 2. Left Inverse for Tall Matrix

```python
T = np.random.randn(6,3)
L = np.linalg.inv(T.T @ T) @ T.T
print('T shape:', T.shape)
print('L shape:', L.shape)
print('L @ T:\n', np.round(L @ T, 6))
```

### Output Interpretation

A full-column-rank tall matrix can have a left inverse. Multiplying from the left returns the identity matrix of size equal to the number of columns.

## 3. Pseudoinverse for Singular Matrix

```python
S = np.array([[1,2], [2,4]])
print('rank:', matrix_rank(S))
Sinv = np.linalg.pinv(S)
print('pseudoinverse:\n', np.round(Sinv, 3))
print('S @ pinv(S):\n', np.round(S @ Sinv, 3))
```

### Output Interpretation

The matrix is singular and has no full inverse, but the pseudoinverse still exists. The product is not the exact identity, but it provides the best generalized inverse behavior.
