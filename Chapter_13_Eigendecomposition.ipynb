## Theoretical Explanation

An eigenvector is a direction that remains on the same line after multiplication by a matrix. The eigenvalue tells how much the vector is scaled. Eigendecomposition factors a square matrix into eigenvectors and eigenvalues. Symmetric matrices are especially important because they have real eigenvalues and orthogonal eigenvectors.

Eigenvalues reveal matrix properties such as determinant, trace, rank, and definiteness. These ideas are central to PCA, optimization, and stability analysis.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import norm, matrix_rank

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 1. Eigenvalues and Eigenvectors

```python
A = np.array([[3,1],[1,3]])
vals, vecs = np.linalg.eig(A)
print('eigenvalues:', vals)
print('eigenvectors:\n', np.round(vecs,3))
for i in range(2):
    print('A v vs lambda v:', np.round(A@vecs[:,i],3), np.round(vals[i]*vecs[:,i],3))
```

### Output Interpretation

For each eigenvector, multiplying by the matrix gives the same result as scaling by the corresponding eigenvalue. Direction is preserved.

## 2. Diagonalization Check

```python
V = vecs
D = np.diag(vals)
A_recon = V @ D @ np.linalg.inv(V)
print('reconstruction error:', norm(A-A_recon))
print('trace:', np.trace(A), 'sum eigenvalues:', np.sum(vals))
print('determinant:', np.linalg.det(A), 'product eigenvalues:', np.prod(vals))
```

### Output Interpretation

The reconstruction verifies diagonalization. The trace equals the sum of eigenvalues, and the determinant equals their product.

## 3. Positive Definiteness

```python
B = np.random.randn(4,3)
S = B.T @ B
vals = np.linalg.eigvalsh(S)
print('eigenvalues of B.T @ B:', np.round(vals,6))
print('positive semidefinite?', np.all(vals >= -1e-10))
```

### Output Interpretation

Matrices of the form `B.T @ B` are positive semidefinite because quadratic forms become squared norms, which cannot be negative.
