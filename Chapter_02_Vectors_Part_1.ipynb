## Theoretical Explanation

A vector is an ordered list of numbers with both algebraic and geometric interpretations. Algebraically, vectors store values. Geometrically, vectors have magnitude and direction. Operations such as addition, scalar multiplication, and dot products reveal relationships between vectors.

The dot product is especially important because it measures alignment. If two vectors are orthogonal, their dot product is zero. This idea supports projections, least squares, correlations, filtering, and many decomposition algorithms.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import norm, matrix_rank

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 1. Vector Orientation and Norms

```python
row = np.array([[1, 2, 3]])
col = np.array([[1], [2], [3]])
flat = np.array([1, 2, 3])

print('row shape:', row.shape)
print('col shape:', col.shape)
print('flat shape:', flat.shape)
print('norm of flat:', norm(flat))
unit = flat / norm(flat)
print('unit vector:', np.round(unit, 3))
print('unit norm:', norm(unit))
```

### Output Interpretation

The row vector, column vector, and flat NumPy array contain the same numbers but have different shapes. Shape matters because some linear algebra operations depend on orientation.

## 2. Dot Product and Orthogonality

```python
a = np.array([2, 3])
b = np.array([3, -2])
print('dot(a,b):', np.dot(a,b))
print('Are a and b orthogonal?', np.isclose(np.dot(a,b), 0))

plt.arrow(0,0,a[0],a[1],head_width=.15,length_includes_head=True,label='a')
plt.arrow(0,0,b[0],b[1],head_width=.15,length_includes_head=True,color='orange',label='b')
plt.xlim(-3,4); plt.ylim(-3,4); plt.grid(True); plt.title('Orthogonal vectors')
plt.show()
```

### Output Interpretation

The zero dot product confirms orthogonality. The plot gives the geometric interpretation: the vectors meet at a right angle.

## 3. Orthogonal Vector Decomposition

```python
t = np.array([3, 4])
r = np.array([4, 1])
parallel = r * (np.dot(t, r) / np.dot(r, r))
perpendicular = t - parallel
print('target:', t)
print('parallel component:', np.round(parallel, 3))
print('perpendicular component:', np.round(perpendicular, 3))
print('sum check:', np.round(parallel + perpendicular, 3))
print('orthogonality check:', np.dot(perpendicular, r))
```

### Output Interpretation

The target vector is decomposed into a component parallel to the reference vector and a residual component perpendicular to the reference vector. This projection idea is central to least squares.
