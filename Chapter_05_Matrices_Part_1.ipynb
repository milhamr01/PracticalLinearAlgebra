## Theoretical Explanation

A matrix can be interpreted as a data table, a collection of column vectors, a collection of row vectors, or a transformation. Element-wise operations behave intuitively, but standard matrix multiplication is based on dot products between rows and columns.

Matrix multiplication is not generally commutative. The inner dimensions must match, and the resulting shape is determined by the outer dimensions. Symmetric matrices are important because many data matrices such as covariance matrices are symmetric.

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import norm, matrix_rank

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 1. Matrix Creation and Special Matrices

```python
A = np.arange(12).reshape(3,4)
I = np.eye(4)
D = np.diag([1,2,3,4])
Z = np.zeros((3,4))
print('A:\n', A)
print('Identity:\n', I)
print('Diagonal:\n', D)
print('Zeros shape:', Z.shape)
```

### Output Interpretation

Special matrices appear frequently in algorithms. Identity matrices preserve values under multiplication, diagonal matrices scale axes, and zero matrices represent additive identity.

## 2. Matrix Multiplication and Shape Rules

```python
B = np.arange(8).reshape(4,2)
C = A @ B
print('A shape:', A.shape)
print('B shape:', B.shape)
print('A @ B shape:', C.shape)
print('A @ B:\n', C)
```

### Output Interpretation

The inner dimensions 4 and 4 match, so multiplication is valid. The product has shape 3 by 2, determined by the outer dimensions.

## 3. Symmetric Matrix from A.T @ A

```python
R = np.random.randn(5,3)
S = R.T @ R
print('S shape:', S.shape)
print('Is symmetric?', np.allclose(S, S.T))
plt.imshow(S, cmap='gray')
plt.title('Symmetric matrix R.T @ R')
plt.colorbar(); plt.show()
```

### Output Interpretation

Multiplying a matrix by its transpose creates a square symmetric matrix. This construction is central to covariance matrices, normal equations, and SVD.
